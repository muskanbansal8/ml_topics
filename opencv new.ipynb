```python
import cv2

cap=cv2.VideoCapture(0)
plugin=cv2.VideoWriter_fourcc(*'XVID')
# saving video width hight
output=cv2.VideoWriter('class.avi',plugin,30,(640,480))

while cap.isOpened() :
    status,img=cap.read()
    #converting image frame into grayscale
    gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
    #cv2.imshow('image1',gray_img)
    cv2.line(img,(0,0),(200,300),(0,100,255),3)
    cv2.rectangle(img,(50,50),(300,300),(0,0,255),3)
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img,'muskan',(150,150),font,1,(50,10,5),5,cv2.LINE_AA)
    # circle
    cv2.circle(img,(200,300),50,(100,150,200),3)
    output.write(img)
    cv2.imshow('image',img)
    cv2.imwrite('pic.jpg',img)
    
    if cv2.waitKey(10) & 0xff == ord('q'):
        
        break
cv2.destroyWindow('image')
cap.release() # to close camera
```

```python
import cv2
#starting camera 
cap=cv2.VideoCapture(0) # we can give video or url in replace of 0
tap1=cap.read()[1]
tap2=cap.read()[1]
tap3=cap.read()[1]# take1
#to make more perfect
gray1=cv2.cvtColor(tap1,cv2.COLOR_BGR2GRAY)
gray2=cv2.cvtColor(tap2,cv2.COLOR_BGR2GRAY)
gray3=cv2.cvtColor(tap3,cv2.COLOR_BGR2GRAY)
# now creating diffrence
def img_diff(x,y,z):
    #diff between x and y -->d1
    d1=cv2.absdiff(x,y)
    #diff between y and z -->d2
    d2=cv2.absdiff(y,z)
    #abs diff d1-d2
    finalimg=cv2.bitwise_and(d1,d2)
    return finalimg
while cap.isOpened():
    status,frame=cap.read()
    motionimg=img_diff(gray1,gray2,gray3)
    #relacing image frame
    gray1=gray2
    gray2=gray3
    gray3=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) # passing live image to gray3 
    cv2.imshow('motion',frame)
    cv2.imshow('motion',motionimg)
    if cv2.waitKey(10) & 0xff ==ord('q'):
        break
        
cv2.destroyAllWindows()
cap.release()        

    
```

```python
#face detection
import cv2
#calling classsifier 
casclf=cv2.CascadeClassifier('face_data.xml')
casclf1=cv2.CascadeClassifier('eye_data.xml')
#casclf2=cv2.CascadeClassifier('smile_data.xml')
cap=cv2.VideoCapture(0)
while cap.isOpened():
    status,frame=cap.read()
    
    #now we can apply classifier in live frame
    #smile=casclf2.detectMultiScale(frame,1.13,5)
    face=casclf.detectMultiScale(frame,1.13,5)
    eye=casclf1.detectMultiScale(frame,1.13,5)# classifier tunning parameter
    print(face)
    for x,y,h,w in face:
        cv2.rectangle(frame,(x,y),(x+h,y+w),(255,0,0),2)
    for x,y,h,w in eye:
        cv2.rectangle(frame,(x,y),(x+h,y+w),(255,0,0),2)  
    
        #only face
    cv2.imshow('face',frame)
    if cv2.waitKey(10) & 0xff==ord('q'):
        break
cv2.destroyAllWindows()
cap.release()
```
